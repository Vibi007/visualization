## Data Visualization:

```python
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('iris.data.txt',names = ['s_len','s_wid','p_len','p_wid','class'])
```

```python
df.columns
```

#### Numerical variables

```python
numerical_variables = ['s_len','s_wid','p_len','p_wid']
```

#### Categorical Variables

```python
categorical_variables = ['class']
```

###### Numerical Variables are visualised with scatter plot

```python
df.plot(kind = 'scatter',x = 's_len',y = 's_wid')
```

###### Categorical Variables are visualised with bar chart ( taking count)

```python
df['class'].value_counts().plot(kind = 'bar')
```

###### Comparison between numerical & categorical is done using box plot

```python
sns.boxplot(data=df,x = 'class', y = 's_wid')
```

###### For multiple categorical variable vs Numerical variable box plot, we can have sub plots

```python
fig,ax = plt.subplots(2,2,figsize = (14,9))
for var,subplot in zip(numerical_variables,ax.flatten()):
    sns.boxplot(x = 'class',y = var, data = df, ax = subplot )
fig.tight_layout()
```

```python
g = sns.FacetGrid(df,col = 'class')
g.map(plt.scatter,'s_wid','s_len')
```

###### For numerical variables we can make pairplot using seaborn

```python
sns.pairplot(df)
```

```python
sns.jointplot(x = 's_len', y = 's_wid',data = df)
```
